```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
N = 100 # Number of points
m = np.array([0,1,2,3,4,5,6,7]) # Measurements
x = np.linspace(0,m[-1],N) # Domain for sine function
data = np.array([0.784-0.762,0.762-0.755,0.755-0.738,0.738-0.721,0.721-0.720,
                 0.720-0.715,0.715-0.692,0.692-0.662]) # Average displacements in units of wavelength
mu = np.mean(data) # Average of averages
data -= mu
err = 0.01*np.ones([len(m)])
theory = 0.05*np.sin(2*np.pi*x/7) # 1/8 of theoretical prediction in units of wavelength
```

```python
plt.figure(figsize=(12,10))
plt.plot(m,data,label='Measured Values')
plt.plot(x,theory,label='Theoretical Prediction')
plt.errorbar(m,data,yerr=err,xerr=None,ls='none')
plt.grid()
plt.legend()
```

```python
# H0, H1, and H2 are all nested hypotheses

def H0(c,x): # Null hypothesis
    return c

def chi2H0(x,data,err,c): # Chi^2 for Null hypothesis
    return np.sum(pow( (data-H0(c,x) )/err,2))

def H1(c,w,x): # First alternative hypothesis, fixing amplitude at theoretical prediction, w is free param
    return 0.4*np.sin(w*x) + c

def chi2H1(x,data,err,c,w): # Chi^2 for first alternative hypothesis
    return np.sum(pow( (data-H1(c,w,x) )/err,2))

def H2(c,w,A,x): # Second alternative hypothesis, amplitude and w are free params
    return A*np.sin(w*x) + c

def chi2H2(x,data,err,c,w,A): # Chi^2 for first alternative hypothesis
    return np.sum(pow( (data-H2(c,w,A,x) )/err,2))

funcH0 = lambda param: chi2H0(x,data,err,param[0]) # Function to minimize for H0
funcH1 = lambda param: chi2H1(x,data,err,param[0],param[1]) # Function to minimize for H0
funcH2 = lambda param: chi2H2(x,data,err,param[0],param[1],param[2]) # Function to minimize for H0
```
